```python
1). what is multithreading in python? hy is it used? Name the module used to handle threads in python
Answer--Multithreading in Python refers to the ability of a program to execute multiple threads concurrently within a single process.
A thread is the smallest unit of a CPU's execution, and multithreading allows for better utilization of a computer's resources,
especially in cases where a program needs to perform multiple tasks simultaneously or handle multiple I/O-bound operations.

Threads share the same memory space within a process, which means they can easily exchange data and communicate with each other.
Multithreading is used primarily for two purposes:

Concurrency: Multithreading allows a program to achieve concurrency, which means multiple tasks can appear to be executing simultaneously
Parallelism: Although Python's 
Global Interpreter Lock (GIL) restricts true parallel execution of threads due to limitations in CPython's memory management, multithreading can still be useful in certain scenarios

The module used to handle threads in Python is called threading. This module provides a high-level interface for creating and working with threads. 
It includes classes and functions to create and manage threads, handle synchronization, and manage thread-local data. 

2. Why threading module used? Write the use of the following functions
(A) activeCount
(B) currentThread
(C) enumerate

Answer--The threading module in Python is used to create and manage threads within a single process. 
It provides a way to achieve concurrency, where multiple threads appear to execute simultaneously and share the same memory space
A--activeCount():
This function returns the number of Thread objects currently alive. It counts both started and not yet started threads.
import threading

def worker():
    print("Thread is working...")

thread1 = threading.Thread(target=worker)
thread2 = threading.Thread(target=worker)

thread1.start()
thread2.start()

print("Active threads:", threading.activeCount())

B--currentThread():
This function returns the current Thread object, corresponding to the caller's thread of control.
import threading

def print_current_thread():
    current = threading.currentThread()
    print("Current thread:", current.name)

thread1 = threading.Thread(target=print_current_thread, name="Thread 1")
thread2 = threading.Thread(target=print_current_thread, name="Thread 2")

thread1.start()
thread2.start()

thread1.join()
thread2.join()


C--enumerate():
This function returns a list of all Thread objects currently alive. It includes both started and not yet started threads.
import threading

def worker():
    print("Thread is working...")

thread1 = threading.Thread(target=worker)
thread2 = threading.Thread(target=worker)

thread1.start()
thread2.start()

threads = threading.enumerate()
print("Enumerated threads:", threads)

3. Explain the following functions
run
start
join
isAlive()
Answer--run():
The run() method is the entry point for the thread's activity. You can subclass the Thread class and override this method with the code that you want the thread to execute. 
When you start a thread using the start() method, it calls the run() method of that thread.

b--start():
The start() method starts the execution of a thread. It allocates system resources, schedules the thread for execution, and calls the thread's run() method.

c--join():
The join() method blocks the calling thread until the thread on which it's called completes its execution. 
It's used to ensure that the calling thread waits for the target thread to finish before proceeding.
d--join():
The join() method blocks the calling thread until the thread on which it's called completes its execution.
It's used to ensure that the calling thread waits for the target thread to finish before proceeding.

4. rite a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes
Answer--import threading

def print_squares():
    for i in range(1, 6):
        print(f"Square of {i}: {i**2}")

def print_cubes():
    for i in range(1, 6):
        print(f"Cube of {i}: {i**3}")

# Create thread objects
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start the threads
thread1.start()
thread2.start()

# Wait for threads to finish
thread1.join()
thread2.join()

print("Both threads have finished.")


5. State advantages and disadvantages of multithreading
Answer--Advantages:

Concurrency

Resource Sharing

I/O-Bound Tasks

Responsiveness

Lower Overhead

Disadvantages:

Complexity: Writing multithreaded code can be complex and error-prone due to issues like race conditions, deadlocks, and synchronization problems.

Synchronization Overhead: Proper synchronization mechanisms (locks, semaphores, etc.) are needed to prevent race conditions, which can introduce overhead and complexity.

Limited Parallelism: Python's Global Interpreter Lock (GIL) restricts true parallel execution of threads in CPython, making it less suitable for CPU-bound tasks. This can limit the potential performance gains from multithreading.

Debugging: Debugging multithreaded code is more challenging due to the potential for non-determin

Performance Trade-offs

Resource Contention

Portability

Unpredictability





```
